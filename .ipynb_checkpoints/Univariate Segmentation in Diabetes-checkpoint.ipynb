```python
# Import necessary libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Load the dataset
data = pd.read_csv('diabetes_mellitus_study.csv')

# Display first few rows of the dataset
print(data.head())

# Summary statistics for glucose level, BMI, and age
summary_stats = data[['glucose_level', 'bmi', 'age']].describe()
print("Summary statistics for glucose level, BMI, and age:")
print(summary_stats)

# Frequency distribution for outcome
outcome_freq = data['outcome'].value_counts()
print("Frequency distribution of outcome (diabetic vs non-diabetic):")
print(outcome_freq)

# Visualize the glucose levels with a histogram
plt.figure(figsize=(10, 6))
sns.histplot(data['glucose_level'], bins=30, kde=True)
plt.title('Distribution of Glucose Levels')
plt.xlabel('Glucose Level')
plt.ylabel('Frequency')
plt.show()

# Boxplot of BMI segmented by outcome
plt.figure(figsize=(10, 6))
sns.boxplot(x='outcome', y='bmi', data=data)
plt.title('Boxplot of BMI by Outcome')
plt.xlabel('Outcome (0=Non-Diabetic, 1=Diabetic)')
plt.ylabel('BMI')
plt.show()

# Histogram of age distribution
plt.figure(figsize=(10, 6))
sns.histplot(data['age'], bins=30, kde=True, color='green')
plt.title('Age Distribution')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.show()
```
